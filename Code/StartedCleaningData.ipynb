I am just going to rid ourselves of all the data that is not nationwide.

```python
data = data[data.usa_or_hhsregion == 'USA']
```

```python
variants = data.variant.unique()
threshold = 0.005
```

```python
for var in variants:
  data.loc[(data.variant == var) & (data.week_ending == data[ (data.variant == var) & (data.share > threshold)].week_ending.min()),'variant_emerges'] = 1
```

Most of the work is done in that `for` loop.  I think this works...

```python
data = pa.get_dummies(data, columns = ['variant','modeltype','time_interval'])
```
